```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error

# -------------------------------
# Load dataset
# -------------------------------
data = pd.read_excel("New Data.xlsx")

# Features and Target
X = data[["Oxide Thickness", "Gate Length", "Work Function", "Vds", "Vgs"]]
y = data["Id"]

# Ensure numeric and positive
y = pd.to_numeric(y, errors="coerce").fillna(1e-30)
y = y.clip(lower=1e-30)

# Log-transform target for stability
y_log = np.log10(y)

# -------------------------------
# LightGBM Regressor (optimized for speed)
# -------------------------------
lgb_model = LGBMRegressor(
    n_estimators=1000,       # reduced from 3000 to speed up
    learning_rate=0.05,      # faster convergence
    num_leaves=32,           # smaller leaf size for speed
    max_depth=-1,            # auto depth
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
    n_jobs=-1
)
lgb_model.fit(X, y_log)

# -------------------------------
# Training Evaluation
# -------------------------------
y_pred_log = lgb_model.predict(X)
y_pred = 10**y_pred_log  # inverse log-transform

r2_train = r2_score(y, y_pred)
mse_train = mean_squared_error(y, y_pred)
mape_train = mean_absolute_percentage_error(y, y_pred)

print("Training RÂ²:", r2_train)
print("Training MSE:", mse_train)
print("Training MAPE:", mape_train)

# -------------------------------
# Testing Input Prediction
# -------------------------------
oxide_test = 1
gate_test = 40
work_func_test = 4.55
vds_test = 1.0
vgs_test = np.arange(0, 1.505, 0.005)

X_test = pd.DataFrame({
    "Oxide Thickness": [oxide_test]*len(vgs_test),
    "Gate Length": [gate_test]*len(vgs_test),
    "Work Function": [work_func_test]*len(vgs_test),
    "Vds": [vds_test]*len(vgs_test),
    "Vgs": vgs_test
})

pred_test_log = lgb_model.predict(X_test)
pred_test = 10**pred_test_log

# -------------------------------
# Plot Id vs Vgs
# -------------------------------
plt.figure(figsize=(8,6))
plt.plot(vgs_test, pred_test, "go-", label="Predicted Id (LightGBM)")
plt.yscale("log")
plt.xlabel("Vgs (V)")
plt.ylabel("Id (A) [Log Scale]")
plt.title("LightGBM: Predicted Id vs Vgs")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.7)
plt.show()

# -------------------------------
# Save Predictions
# -------------------------------
df_test = pd.DataFrame({
    "Vgs": vgs_test,
    "Predicted_Id": pred_test
})
df_test.to_excel("LightGBM_Testing_Input_Predictions.xlsx", index=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error

# -------------------------------
# Load dataset
# -------------------------------
data = pd.read_excel("New Data.xlsx")

# Features and Target
X = data[["Oxide Thickness", "Gate Length", "Work Function", "Vds", "Vgs"]]
y = data["Id"]

# Ensure numeric and positive
y = pd.to_numeric(y, errors="coerce").fillna(1e-30)
y = y.clip(lower=1e-30)

# Log-transform target for stability
y_log = np.log10(y)

# -------------------------------
# LightGBM Regressor (optimized for speed)
# -------------------------------
lgb_model = LGBMRegressor(
    n_estimators=1000,       # reduced from 3000 to speed up
    learning_rate=0.05,      # faster convergence
    num_leaves=32,           # smaller leaf size for speed
    max_depth=-1,            # auto depth
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
    n_jobs=-1
)
lgb_model.fit(X, y_log)

# -------------------------------
# Training Evaluation
# -------------------------------
y_pred_log = lgb_model.predict(X)
y_pred = 10**y_pred_log  # inverse log-transform

r2_train = r2_score(y, y_pred)
mse_train = mean_squared_error(y, y_pred)
mape_train = mean_absolute_percentage_error(y, y_pred)

print("Training RÂ²:", r2_train)
print("Training MSE:", mse_train)
print("Training MAPE:", mape_train)

# -------------------------------
# Testing Input Prediction
# -------------------------------
oxide_test = 1
gate_test = 45
work_func_test = 4.5
vds_test = 1.0
vgs_test = np.arange(0, 1.505, 0.005)

X_test = pd.DataFrame({
    "Oxide Thickness": [oxide_test]*len(vgs_test),
    "Gate Length": [gate_test]*len(vgs_test),
    "Work Function": [work_func_test]*len(vgs_test),
    "Vds": [vds_test]*len(vgs_test),
    "Vgs": vgs_test
})

pred_test_log = lgb_model.predict(X_test)
pred_test = 10**pred_test_log

# -------------------------------
# Plot Id vs Vgs
# -------------------------------
plt.figure(figsize=(8,6))
plt.plot(vgs_test, pred_test, "go-", label="Predicted Id (LightGBM)")
plt.yscale("log")
plt.xlabel("Vgs (V)")
plt.ylabel("Id (A) [Log Scale]")
plt.title("LightGBM: Predicted Id vs Vgs")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.7)
plt.show()

# -------------------------------
# Save Predictions
# -------------------------------
df_test = pd.DataFrame({
    "Vgs": vgs_test,
    "Predicted_Id": pred_test
})
df_test.to_excel("LightGBM_Testing_Input_Predictions_Gate.xlsx", index=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error

# -------------------------------
# Load dataset
# -------------------------------
data = pd.read_excel("New Data.xlsx")

# Features and Target
X = data[["Oxide Thickness", "Gate Length", "Work Function", "Vds", "Vgs"]]
y = data["Id"]

# Ensure numeric and positive
y = pd.to_numeric(y, errors="coerce").fillna(1e-30)
y = y.clip(lower=1e-30)

# Log-transform target for stability
y_log = np.log10(y)

# -------------------------------
# LightGBM Regressor (optimized for speed)
# -------------------------------
lgb_model = LGBMRegressor(
    n_estimators=1000,       # reduced from 3000 to speed up
    learning_rate=0.05,      # faster convergence
    num_leaves=32,           # smaller leaf size for speed
    max_depth=-1,            # auto depth
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
    n_jobs=-1
)
lgb_model.fit(X, y_log)

# -------------------------------
# Training Evaluation
# -------------------------------
y_pred_log = lgb_model.predict(X)
y_pred = 10**y_pred_log  # inverse log-transform

r2_train = r2_score(y, y_pred)
mse_train = mean_squared_error(y, y_pred)
mape_train = mean_absolute_percentage_error(y, y_pred)

print("Training RÂ²:", r2_train)
print("Training MSE:", mse_train)
print("Training MAPE:", mape_train)

# -------------------------------
# Testing Input Prediction
# -------------------------------
oxide_test = 1
gate_test = 40
work_func_test = 4.5
vds_test = 0.75
vgs_test = np.arange(0, 1.505, 0.005)

X_test = pd.DataFrame({
    "Oxide Thickness": [oxide_test]*len(vgs_test),
    "Gate Length": [gate_test]*len(vgs_test),
    "Work Function": [work_func_test]*len(vgs_test),
    "Vds": [vds_test]*len(vgs_test),
    "Vgs": vgs_test
})

pred_test_log = lgb_model.predict(X_test)
pred_test = 10**pred_test_log

# -------------------------------
# Plot Id vs Vgs
# -------------------------------
plt.figure(figsize=(8,6))
plt.plot(vgs_test, pred_test, "go-", label="Predicted Id (LightGBM)")
plt.yscale("log")
plt.xlabel("Vgs (V)")
plt.ylabel("Id (A) [Log Scale]")
plt.title("LightGBM: Predicted Id vs Vgs")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.7)
plt.show()

# -------------------------------
# Save Predictions
# -------------------------------
df_test = pd.DataFrame({
    "Vgs": vgs_test,
    "Predicted_Id": pred_test
})
df_test.to_excel("LightGBM_Testing_Input_Predictions_Vds.xlsx", index=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error

# -------------------------------
# Load dataset
# -------------------------------
data = pd.read_excel("New Data.xlsx")

# Features and Target
X = data[["Oxide Thickness", "Gate Length", "Work Function", "Vds", "Vgs"]]
y = data["Id"]

# Ensure numeric and positive
y = pd.to_numeric(y, errors="coerce").fillna(1e-30)
y = y.clip(lower=1e-30)

# Log-transform target for stability
y_log = np.log10(y)

# -------------------------------
# LightGBM Regressor (optimized for speed)
# -------------------------------
lgb_model = LGBMRegressor(
    n_estimators=1000,       # reduced from 3000 to speed up
    learning_rate=0.05,      # faster convergence
    num_leaves=32,           # smaller leaf size for speed
    max_depth=-1,            # auto depth
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
    n_jobs=-1
)
lgb_model.fit(X, y_log)

# -------------------------------
# Training Evaluation
# -------------------------------
y_pred_log = lgb_model.predict(X)
y_pred = 10**y_pred_log  # inverse log-transform

r2_train = r2_score(y, y_pred)
mse_train = mean_squared_error(y, y_pred)
mape_train = mean_absolute_percentage_error(y, y_pred)

print("Training RÂ²:", r2_train)
print("Training MSE:", mse_train)
print("Training MAPE:", mape_train)

# -------------------------------
# Testing Input Prediction
# -------------------------------
oxide_test = 1
gate_test = 45
work_func_test = 4.55
vds_test = 0.75
vgs_test = np.arange(0, 1.505, 0.005)

X_test = pd.DataFrame({
    "Oxide Thickness": [oxide_test]*len(vgs_test),
    "Gate Length": [gate_test]*len(vgs_test),
    "Work Function": [work_func_test]*len(vgs_test),
    "Vds": [vds_test]*len(vgs_test),
    "Vgs": vgs_test
})

pred_test_log = lgb_model.predict(X_test)
pred_test = 10**pred_test_log

# -------------------------------
# Plot Id vs Vgs
# -------------------------------
plt.figure(figsize=(8,6))
plt.plot(vgs_test, pred_test, "go-", label="Predicted Id (LightGBM)")
plt.yscale("log")
plt.xlabel("Vgs (V)")
plt.ylabel("Id (A) [Log Scale]")
plt.title("LightGBM: Predicted Id vs Vgs")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.7)
plt.show()

# -------------------------------
# Save Predictions
# -------------------------------
df_test = pd.DataFrame({
    "Vgs": vgs_test,
    "Predicted_Id": pred_test
})
df_test.to_excel("LightGBM_Testing_Input_Predictions_Three.xlsx", index=False)
```
