```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error

# -------------------------------
# Load dataset
# -------------------------------
data = pd.read_excel("Final data.xlsx")

# Features and Target
X = data[["Oxide Thickness", "Gate Length", "Work Function", "Vds", "Vgs"]]
y = data["Id"]

# Ensure numeric and positive
y = pd.to_numeric(y, errors="coerce").fillna(1e-30)
y = y.clip(lower=1e-30)

# Log-transform target for stability
y_log = np.log10(y)

# -------------------------------
# Random Forest Regressor
# -------------------------------
rf = RandomForestRegressor(
    n_estimators=2000,        # large number of trees for stability
    max_depth=20,             # limit depth to avoid overfitting
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    random_state=42,
    n_jobs=-1
)
rf.fit(X, y_log)

# -------------------------------
# Training Evaluation
# -------------------------------
y_pred_log = rf.predict(X)
y_pred = 10**y_pred_log  # inverse log-transform

r2_train = r2_score(y, y_pred)
mse_train = mean_squared_error(y, y_pred)
mape_train = mean_absolute_percentage_error(y, y_pred)

print("Training RÂ²:", r2_train)
print("Training MSE:", mse_train)
print("Training MAPE:", mape_train)

# -------------------------------
# Testing Input Prediction
# -------------------------------
oxide_test = 1
gate_test = 45.5
work_func_test = 4.5
vds_test = 1.0
vgs_test = np.arange(0, 1.505, 0.005)

X_test = pd.DataFrame({
    "Oxide Thickness": [oxide_test]*len(vgs_test),
    "Gate Length": [gate_test]*len(vgs_test),
    "Work Function": [work_func_test]*len(vgs_test),
    "Vds": [vds_test]*len(vgs_test),
    "Vgs": vgs_test
})

pred_test_log = rf.predict(X_test)
pred_test = 10**pred_test_log

# -------------------------------
# Plot Id vs Vgs
# -------------------------------
plt.figure(figsize=(8,6))
plt.plot(vgs_test, pred_test, "ms-", label="Predicted Id (Random Forest)")
plt.yscale("log")
plt.xlabel("Vgs (V)")
plt.ylabel("Id (A) [Log Scale]")
plt.title("Random Forest: Predicted Id vs Vgs")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.7)
plt.show()

# -------------------------------
# Save Predictions
# -------------------------------
df_test = pd.DataFrame({
    "Vgs": vgs_test,
    "Predicted_Id": pred_test
})
df_test.to_excel("RF_Gate_vary.xlsx", index=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error

# -------------------------------
# Load dataset
# -------------------------------
data = pd.read_excel("New Data.xlsx")

# Features and Target
X = data[["Oxide Thickness", "Gate Length", "Work Function", "Vds", "Vgs"]]
y = data["Id"]

# Ensure numeric and positive
y = pd.to_numeric(y, errors="coerce").fillna(1e-30)
y = y.clip(lower=1e-30)

# Log-transform target for stability
y_log = np.log10(y)

# -------------------------------
# Random Forest Regressor
# -------------------------------
rf = RandomForestRegressor(
    n_estimators=2000,        # large number of trees for stability
    max_depth=20,             # limit depth to avoid overfitting
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    random_state=42,
    n_jobs=-1
)
rf.fit(X, y_log)

# -------------------------------
# Training Evaluation
# -------------------------------
y_pred_log = rf.predict(X)
y_pred = 10**y_pred_log  # inverse log-transform

r2_train = r2_score(y, y_pred)
mse_train = mean_squared_error(y, y_pred)
mape_train = mean_absolute_percentage_error(y, y_pred)

print("Training RÂ²:", r2_train)
print("Training MSE:", mse_train)
print("Training MAPE:", mape_train)

# -------------------------------
# Testing Input Prediction
# -------------------------------
oxide_test = 1
gate_test = 45
work_func_test = 4.55
vds_test = 0.75
vgs_test = np.arange(0, 1.505, 0.005)

X_test = pd.DataFrame({
    "Oxide Thickness": [oxide_test]*len(vgs_test),
    "Gate Length": [gate_test]*len(vgs_test),
    "Work Function": [work_func_test]*len(vgs_test),
    "Vds": [vds_test]*len(vgs_test),
    "Vgs": vgs_test
})

pred_test_log = rf.predict(X_test)
pred_test = 10**pred_test_log

# -------------------------------
# Plot Id vs Vgs
# -------------------------------
plt.figure(figsize=(8,6))
plt.plot(vgs_test, pred_test, "ms-", label="Predicted Id (Random Forest)")
plt.yscale("log")
plt.xlabel("Vgs (V)")
plt.ylabel("Id (A) [Log Scale]")
plt.title("Random Forest: Predicted Id vs Vgs")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.7)
plt.show()

# -------------------------------
# Save Predictions
# -------------------------------
df_test = pd.DataFrame({
    "Vgs": vgs_test,
    "Predicted_Id": pred_test
})
df_test.to_excel("RandomForest_Testing_Input_Predictions_Three.xlsx", index=False)
```
